```python
def delays_information():
    location = input("Insert the Icao of the airport of your choice (4 letters) : ")
    url = 'https://aerodatabox.p.rapidapi.com/airports/icao/' + location + '/delays/2022-06-03T12:00/2022-06-04T00:00'


    headers = {
        'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
        'x-rapidapi-key': "9613ca6ca5msh62659872159916fp1a0a2cjsndd9a3e6ff324"
        }

    response = requests.get(url, headers=headers)
    df = response.json()
    df = pd.json_normalize(df)
    df['departure_day'] = df['fromUtc'].str[:10]
    return df
    
def departure_date():
    info = delays_information()
    date = input("Insert the date of your departure under this format --> YYYY-MM-DD : ")
    df = info[info['departure_day']==date]
    return df
    
    
```
